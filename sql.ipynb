## Sentencia select
Esta es la consulta m치s b치sica y su estructura es muy simple:

Se define lo que se quiere ver despu칠s del SELECT

# 6.3 Operadores l칩gicos en WHERE췄
La sentencia WHERE puede combinarse con los operadores l칩gicos: AND, OR y NOT. Los operadores AND y OR se utilizan para filtrar registros basados en m치s de una condici칩n.

El operador AND muestra un registro si todas las condiciones separadas por AND son VERDADERA(s).
El operador OR muestra un registro si alguna de las condiciones separadas por OR es VERDADERA(s).
El operador NOT muestra un registro si la(s) condici칩n(es) NO es VERDADERA(s).
Sintaxis deAND:
SELECT column1, column2, ...

FROM table_name

WHERE condition1 AND condition2 AND condition3 ...;

1 Sintaxis de OR
SELECT column1, column2, ...

FROM table_name

WHERE condition1 OR condition2 ...;

1 Sintaxis de NOT
SELECT column1, column2, ...

FROM table_name

WHERE NOT condition;

# 6.4 Otros operadores que trabajan con la sentencia WHERE췄
Los siguientes operadores tambi칠n pueden ser utilizados en la sentencia WHERE:

| = | Igual

| > | Mayor que

| < | Menor que

| >= | Mayor o igual que

| <= | Menor o igual que

| <> | No igual. Nota: En algunas versiones de SQL este operador puede ser escrito como !=

| BETWEEN | Entre cierto rango

| LIKE | Para consultar los datos basados en informaci칩n parcial, se utiliza el operador LIKE en la sentencia WHERE. Busca un patr칩n. El patr칩n va antes, despu칠s o entre un par de % (wildcard). Ejemplo: LIKE 's%' busca cualquier texto que comience con 's'. LIKE '%er' busca cualquier texto que termine con 'er'. Y LIKE '%per%' busca cualquier texto que contenga 'per'.

| IN | Para especificar varios valores posibles para una columna o subquery. Ejemplo de sintaxis cuando se especifica varios valores:

SELECT column1
FROM Table_name
WHERE column1 IN (value1, value2);.

Ejemplo se sintaxis cuando se usa IN en una subquery:

SELECT TrackId, Name, AlbumId
FROM Tracks
WHERE AlbumId IN
    (SELECT AlbumId
    FROM Albums
    WHERE ArtistId = 12);.
